# **Word Problem**

A health bar is made from a corn mixture, an egg mixture, and a vegetable mixture. One ounce of the corn mixture contains 0.25 grams of protein, 0.3 grams of carbohydrates, and 0.4 grams of fat. The egg mixture contains 0.4 grams of protein, and 0.2 grams of bot carbohydrates and fat. Lastly, the vegetable mixture contains 0.2 grams of protein, 0.1 grams of carbohydrates, and 0.3 grams of fat. How many ounces of each mixture should be used to create a health bar that contains 22 grams of protein, 18 grams of carbohydrates, and 28 grams of fat?

Source: [Youtube Link](https://youtu.be/LGdJXDjhZh4)



$$ A = \left\{ \begin{array}\\ 0.25x+0.4y+0.2z=22\\ 0.3x+0.2y+0.1z=18 \\ 0.4x+0.2y+0.3z=28 \end{array} \right. $$

$$
\begin{bmatrix} 
0.25 & 0.4 & 0.2 \\
0.3 & 0.2 & 0.1\\
0.4 & 0.2 & 0.3\\
\end{bmatrix}
\begin{bmatrix} 
x\\
y\\
z\\
\end{bmatrix} = 
\begin{bmatrix} 
22\\
18\\
28\\
\end{bmatrix}
$$

```python
import numpy as np
import matplotlib.pyplot as plt
```

# **Declaration of the Array/Matrix**

```python
A = np.array([[0.25,0.4,0.2],[0.3,0.2,0.1],[0.4,0.2,0.3]])
A
```

# **Visualization of the Vectors**

```python
fig = plt.figure()
ax = fig.gca(projection='3d')
ax.set_xlim([0, 0.5])
ax.set_ylim([0, 0.5])
ax.set_zlim([0, 0.5])

ax.quiver(0, 0, 0, A[0,0], A[0,1], A[0,2], arrow_length_ratio=0.05, colors='red')
ax.quiver(0, 0, 0, A[1,0], A[1,1], A[1,2], arrow_length_ratio=0.05, colors='blue')
ax.quiver(0, 0, 0, A[2,0], A[2,1], A[2,2], arrow_length_ratio=0.05, colors='green')
plt.show()
```

# **Solving the System of Linear Equations**

```python
A = np.matrix([[0.25,0.4,0.2],[0.3,0.2,0.1],[0.4,0.2,0.3]])
B = np.matrix([[22],[18],[28]])

A_inverse = np.linalg.inv(A)

X = A_inverse * B
print("x = ", X[0])
print("y = ", X[1])
print("z = ", X[2])
```

The values above which are 40, 15, and 30 are the values of x, y, and z respectively.
