```
import numpy as np
import matplotlib.pyplot as plt
```

#Equations

```
## Polynomials Equations
def b1(x): return x**3 - 4*x**2 + 2*x + 3     
def b1_prime(x): return 3*x - 8*x + 2  

## Trigonometric Equations
def e(x): return np.cos(x) + np.sin(x) 
def e_prime(x): return -np.sin(x) + np.cos(x)

## Logarithmic Equations
def k(x): return np.log(x-1) * (2*x - 5)
def k_prime(x): return ((2*x - 5)/(x - 1)) + 2*np.log(x-1)
```

#Brute Force

### Using Polynomial Function


```
history_x = []
history_x_prime = []
## Solving for a single root
epochs = 10
x=0
root = None
for epoch in range(epochs):
  x_prime = b1(x)
  history_x.append(x)
  history_x_prime.append(x_prime)
  if np.allclose(x,x_prime):
    #root = x
    break
  x = x_prime
```

```
history_x = history_x[0:]
history_x_prime = history_x_prime[0:]
x_range = np.arange(len(history_x),dtype=int)

plt.figure(figsize=(10,7))
plt.plot(x_range, history_x, 'o--',color='dodgerblue', label='x')
plt.plot(x_range, history_x_prime,'D--', color='orange', label='x\'')

plt.title("Fig. 2.1 Convergent Solution")
plt.xlabel("Epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1, 1.0))
plt.legend()
plt.grid()
plt.show()
```

#Newton Raphson Method

### Using Polynomial Function

```
history_x = []
history_x_prime = []

## Single Root
epochs = 10
x = 0
root = None
for epoch in range(epochs):
  x_prime = x - (b1(x)/b1_prime(x))
  history_x.append(x)
  history_x_prime.append(x_prime)
  if np.allclose(x, x_prime):
    root = x
    break
  x = x_prime
```

#Bisection Method

### Using Polynomial Function

```
history_x = []
history_x_prime = []

# Single Root
i1, i2 = 1.1, 2
y1, y2 = b1(i1), b1(i2)
root = None
end_bisect = 0
if np.sign(y1) == np.sign(y2):
    print("Root cannot be found in the given interval")
else:
    for bisect in range(0,10):
        midp = np.mean([i1,i2])
        y_mid = b1(midp)
        #y1 = f(i1)
        history_x.append(i1)
        history_x_prime.append(midp)
        if np.allclose(0,y1, 1e-06):
            root = i1
            end_bisect = bisect
            break
        if np.sign(y1) != np.sign(y_mid): #root is in first-half interval
            i2 = midp
        else: #root is in second-half interval
            i1 = midp
```

```
history_x = history_x[0:]
history_x_prime = history_x_prime[0:]
x_range = np.arange(len(history_x),dtype=int)

plt.figure(figsize=(10,7))
plt.plot(x_range, history_x, 'o--',color='dodgerblue', label='x')
plt.plot(x_range, history_x_prime,'D--', color='orange', label='x\'')

plt.title("Fig. 2.1 Convergent Solution")
plt.xlabel("Epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1, 1.0))
plt.legend()
plt.grid()
plt.show()
```

#Regula Falsi

### Using Polynomial Function

```
history_x = []
history_x_prime = []

a,b = 1,2
y1, y2 = b1(a), b1(b)
root = None 
pos = 0
if np.allclose(0,y1): root = a
elif np.allclose(0,y2): root = b
elif np.sign(y1) == np.sign(y2):
  print("No root here")
else:
  for pos in range(0,10):
    c = b - (b1(b)*(b-a))/(b1(b)-b1(a)) ##false root
    x_prime = b1(c)
    history_x.append(b)
    history_x_prime.append(c)
    if np.allclose(0,b1(c), 1e-06):
      #root = c
      break
    if np.sign(b1(a)) != np.sign(b1(c)): b,y2 = c,b1(c)
    else: a,y1 = c,b1(c) 
```

```
history_x = history_x[0:]
history_x_prime = history_x_prime[0:]
x_range = np.arange(len(history_x),dtype=int)

plt.figure(figsize=(10,7))
plt.plot(x_range, history_x, 'o--',color='dodgerblue', label='x')
plt.plot(x_range, history_x_prime,'D--', color='orange', label='x\'')

plt.title("Fig. 2.1 Convergent Solution")
plt.xlabel("Epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1, 1.0))
plt.legend()
plt.grid()
plt.show()
```

# Secant Method 

### Using Polynomial Function

```
history_x = []
history_x_prime = []

x0, x1 = 2, 4
epochs = 10
root = None
for epoch in range(epochs):
  x2 = x1 - (b1(x1)*(x1-x0))/(b1(x1)-b1(x0))
  history_x.append(x1)
  history_x_prime.append(x2)
  if np.allclose(x1,x2): 
    root = x2
    break
  else:
    x0,x1 = x1,x2
```

```
history_x = history_x[0:]
history_x_prime = history_x_prime[0:]
x_range = np.arange(len(history_x),dtype=int)

plt.figure(figsize=(10,7))
plt.plot(x_range, history_x, 'o--',color='dodgerblue', label='x')
plt.plot(x_range, history_x_prime,'D--', color='orange', label='x')

plt.title("Fig. 2.1 Convergent Solution")
plt.xlabel("Epochs")
plt.xticks(np.arange(min(x_range), max(x_range)+1, 1.0))
plt.legend()
plt.grid()
plt.show()
```
